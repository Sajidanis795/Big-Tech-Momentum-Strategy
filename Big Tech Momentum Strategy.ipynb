```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# List of tech stocks
stocks = ['META', 'AMZN', 'AAPL', 'MSFT', 'GOOGL', 'PYPL', 'ADBE', 'NVDA']

# Download historical data
data = yf.download(stocks, start=(datetime.now() - timedelta(days=365)), end=datetime.now())

# Calculate returns over the past month and format as percentage
returns = (data['Adj Close'].pct_change(30).iloc[-1]*100).round(2)

# Get the two stocks with the highest returns
top_stocks = returns.nlargest(2)

print(top_stocks)


# Plotting
plt.figure(figsize=(12, 8))
returns.plot(kind='bar')
plt.title('One Month Returns')
plt.ylabel('Return (%)')
plt.show()
```
